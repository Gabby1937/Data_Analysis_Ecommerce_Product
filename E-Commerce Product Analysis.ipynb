```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("C:/Users/Gabriel Johnson/Downloads/Ecommerce_Product_Store.csv", encoding='latin')
df.head(20)
```

```python
regional_sales = df.groupby('Region').sum()
```

```python
labels = regional_sales.Sales_Profit.index
values = regional_sales.Sales_Profit
```

```python
plt.figure(figsize=(10, 3))
plt.bar(labels, values, color='blue')
plt.xlabel('Sales Profit')
plt.ylabel('Region')
plt.title("PROFIT ON REGION")
plt.show()
```

```python
plt.pie(values, labels=labels, colors=['red', 'blue', 'green', 'purple', 'orange'], autopct="%1.1f%%", shadow=True, startangle=90)
plt.title("Total Regional Profit Division")
plt.show()
```

```python
df.groupby("Categories").sum()
```

```python
product_categories = df.groupby("Categories").sum()
categories = product_categories.Sales_Profit.index
total_profit = product_categories.Sales_Profit
```

```python
plt.figure(figsize=(15, 3))

plt.bar(categories, total_profit, color = 'green')
plt.xlabel('Product Categories')
plt.xticks(rotation=45)
plt.ylabel('Total Profit')
plt.title("Total Profit on Categories")
plt.show()
```

```python
# Tax charges on each categories
Total_tax_charges = df.groupby("Categories").sum()
categories = Total_tax_charges.Tax_charges.index
tax_charges = Total_tax_charges.Tax_charges

Total_tax_charges
```

```python
plt.figure(figsize=(15, 5))
plt.bar(categories, tax_charges, color='red')
plt.xticks(rotation=45)
plt.xlabel("Categories")
plt.ylabel("Tax Charges\n")
plt.title("Tax Charges on Product Categories\n")
plt.show()
```

```python
plt.figure(figsize=(15, 5))
plt.pie(tax_charges, labels=categories, autopct="%1.1f%%", shadow=True, colors=['red', 'blue', 'green', 'purple', 'orange', 'pink', 'cyan', 
                                        'yellow', 'brown', 'red', 'blue', 'green', 'purple', 
                                        'orange', 'pink', 'cyan', 'yellow'])
plt.xticks(rotation=45)
plt.title("Tax Charges on Product Categories\n")
plt.show()
```
